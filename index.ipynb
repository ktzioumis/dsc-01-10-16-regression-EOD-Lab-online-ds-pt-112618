### Let's do Regression 

Right here is the real deal. Let's perform a number of simple regression experiments between the chosen independent variables and the dependent variable (price). We shall do this is a loop and in every iteration, we shall pick one of the independent variables  perform following steps:

* Run a simple OLS regression between independent and dependent variables
* Plot a regression line on the scatter plots
* Plot the residuals using `sm.graphics.plot_regress_exog()`.
* Plot a Q-Q plot for regression residuals normality test 
* Store following values in array for each iteration:
    * Independent Variable
    * r_squared'
    * intercept'
    * 'slope'
    * 'p-value'
    * 'normality (JB)' 
* Comment on each output 

```python
ind_vars
```

```python

ind_vars=list(bost.columns)
ind_vars.pop()
ind_vars
```

```python
# Your code here
vals=[]
for fn in ind_vars:
    fmod=f'medv~{fn}'
    model=ols(formula=fmod,data=bost).fit()
    print("regressions analysis for independant variable:",fn)
#     print(model.summary())
    x_new=pd.DataFrame({fn:[bost[fn].min(),bost[fn].max()]})
    ypred=model.predict(x_new)
    plt.scatter(bost[fn],bost['medv'],label=fn,color='blue')
    plt.plot(x_new,ypred,color='red',label='regression line')
    plt.legend()
    plt.title(fn)
    plt.show()
    fig = plt.figure(figsize=(15,8))
    sm.graphics.plot_regress_exog(model, fn,fig=fig)
    plt.show()
    residuals = model.resid
    fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True, label=fn)
    plt.legend()
    plt.show()
    vals.append([fn,model.rsquared,model.params[0],model.params[1],model.pvalues[1],sms.jarque_bera(model.resid)[0]])
    print(fn,'R-squared',model.rsquared,'intercept',model.params[0],'slope',model.params[1],'p-value',model.pvalues[1],'normality(JB)',sms.jarque_bera(model.resid)[0],'\n')
```

```python
pd.DataFrame(vals, columns=['ind-var','R-sq','intercept','slope','p-val','norm(JB)'])
```
